```python
#read the tsv file
df = pd.read_csv("idmapping_2024_03_28.tsv", sep="\t")
```

```python
df
```

```python
colum_to_keep = ["From", "Entry","Entry Name",	"Gene Names", "Protein names", "Length", "Sequence"]
```

```python
# only keep the columns we need
df = df[colum_to_keep]
```

```python
df.head()
```

```python
# remove row with duplicated Sequence col
df = df.drop_duplicates(subset="Sequence", ignore_index=True)
df.index
```

```python
df.duplicated(subset="Sequence").sum()
```

```python
df
```

```python
# write in "adam_seq.fasta" the sequence name and sequence
with open("satijas_genes_seq.fasta", "w") as f:
    for i in df["From"].values:
        f.write(">"+ i + " " + df[df["From"] == i]["Entry"].values[0] + " " + df[df["From"] == i]["Entry Name"].values[0] + " " + str(df[df["From"] == i].index) +"\n")
        f.write(str(df[df["From"] == i]["Sequence"].values[0]) + "\n")
```

```python
# write fasta file by looping over the dataframe rows
with open("zebra_fish.fasta", "w") as f:
    for i in range(len(df)):
        f.write(">"+ "from_"+df["From"][i] + "_entry_" + df["Entry"][i] + "_gene_names_" + df["Gene Names"][i].replace(" ", "_").replace("/","").replace(":","_") + "_" + str(i) +"\n")
        f.write(df["Sequence"][i] + "\n")
```

```python
# write fasta file by looping over the dataframe rows
with open("norman.fasta", "w") as f:
    for i in range(len(df)):
        f.write(">"+ "from_"+df["From"][i] + "_entry_" + df["Entry"][i] + "_gene_names_" + df["Gene Names"][i].replace(" ", "_").replace("/","").replace(":","_") + "_" + str(i) +"\n")
        f.write(df["Sequence"][i] + "\n")
```
